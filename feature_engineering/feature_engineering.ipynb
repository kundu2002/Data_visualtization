```python
y=df['SalePrice']
X=df.drop(columns=['SalePrice', 'Id'])
```

```python
num = X.select_dtypes(include=['number']).columns
cat = X.select_dtypes(include=['object']).columns
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer

# Assuming X is your DataFrame
num = X.select_dtypes(include=['number']).columns
cat = X.select_dtypes(include=['object']).columns

# Impute missing values in numerical columns with the median
imputer_num = SimpleImputer(strategy='median')
X[num] = imputer_num.fit_transform(X[num])

# Alternatively, impute missing values in numerical columns with the mean
# X[num] = X[num].fillna(X[num].mean())

# Impute missing values in categorical columns with the most frequent value
imputer_cat = SimpleImputer(strategy='most_frequent')
X[cat] = imputer_cat.fit_transform(X[cat])
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize numerical features
scaler = StandardScaler()
X[num] = scaler.fit_transform(X[num])
```

```python
# One-hot encode categorical features
X = pd.get_dummies(X, columns=cat)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```
